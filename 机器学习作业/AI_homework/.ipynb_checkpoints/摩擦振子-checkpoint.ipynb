```python
# -*- coding: utf-8 -*-
"""
Created on Thu Apr  4 19:35:00 2019

@author: LGF
"""
from scipy.optimize import curve_fit
# from scipy import *
# from matplotlib.pyplot import *
import scipy.stats as stats
import matplotlib.pyplot as plt
import numpy as np
plt.rcParams['axes.unicode_minus'] = False
a1 = np.loadtxt("data.txt")
b1 = np.copy(a1.T)
x = b1[0, :]
y = b1[1, :]
def func(x, a, b, c):
    return a*np.cos(b * x + c)


popt, pcov = curve_fit(func, x, y, maxfev=5000)
print(popt)
xx = np.linspace(min(x), max(x), 200)
a = popt[0]
b = popt[1]
c = popt[2]
yvals = func(xx, a, b, c)
r2 = stats.pearsonr(x, y)
print('r 等于', r2[0])
print('popt:', popt)
print('系数a:', a)
print('系数b:', b)
print('系数c:', c)
print('系数pcov:', pcov)
# 绘图
plt.figure(dpi=300)
a = round(a, 2)
b = round(b, 2)
c = round(c, 2)
plot1 = plt.plot(x, y, '*', label='original values')
plot2 = plt.plot(xx, yvals, 'r')
plt.rcParams['savefig.dpi'] = 300
plt.rcParams['figure.dpi'] = 300
plt.xlabel('Time/s')
plt.ylabel('X/m')
plt.legend(loc='upper left', fontsize='small')
plt.show()
```

```python
# -*- coding: utf-8 -*-
"""
Created on Thu Apr  4 19:35:00 2019

@author: LGF
"""
from scipy.optimize import curve_fit
# from scipy import *
# from matplotlib.pyplot import *
import scipy.stats as stats
import matplotlib.pyplot as plt
import numpy as np
plt.rcParams['axes.unicode_minus'] = False
a1 = np.loadtxt("data.txt")
b1 = np.copy(a1.T)
x = b1[0, :]
y = b1[1, :]

def func(x, a, b, c, d, f):
    return a * np.exp(b * x) * np.cos(c * x + d) + f


popt, pcov = curve_fit(func, x, y, maxfev=5000)
print(popt)
a = popt[0]
b = popt[1]
c = popt[2]
d = popt[3]
f = popt[4]
xx = np.linspace(min(x), max(x), 200)
yvals = func(xx, a, b, c, d, f)
r2 = stats.pearsonr(x, y)
print('r 等于', r2[0])
print('popt:', popt)
print('系数a:', a)
print('系数b:', b)
print('系数c:', c)
print('系数d:', d)
print('系数f:', f)
print('系数pcov:', pcov)
print('系数yvals:', yvals)
# 绘图
plt.figure(dpi=300)
a = round(a, 2)
b = round(b, 2)
c = round(c, 2)
d = round(d, 2)
f = round(f, 2)
M = str(a) + 'exp(' + str(b) + 'x)' + 'cos(' + str(c) + 'x' + '+' + str(d) + ')' + '+' + str(f)

plot1 = plt.plot(x, y, '*', label='original values')
plot2 = plt.plot(xx, yvals, 'r', label=M)
plt.rcParams['savefig.dpi'] = 300
plt.rcParams['figure.dpi'] = 300
plt.xlabel('Time/s')
plt.ylabel('X/m')
plt.legend(loc='upper left', fontsize='small')
plt.show()
```
